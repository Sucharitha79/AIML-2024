```python
people=['Ann','Brandon','Chen','David','Emily','Farook']
age=[21,12,32,45,37,18]
weight=[ 55,35,77,68,70,60]
height=[160,135,170,165,173,168]
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, width=bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, width=bar_width, color='red', label='Weight')
plt.title('Bar Chart - Health Care Data')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()
```
